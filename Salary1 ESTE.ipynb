```python
#columna salary y salary_currency se eliminan al estar en la de salary in usd 

df.drop(columns=['salary','salary_currency'], inplace=True)

df
```

```python
# Se investigan las categorías de las columnas categóricas

print(df.experience_level.value_counts())  #Cuatro categorias

print(df.employment_type.value_counts())   #Cuatro categorias

print(df.employee_residence.value_counts())  #Demasiadas, las dejaremos en ocho + other

print(df.company_location.value_counts())  #Demasiadas, las dejaremos en ocho + other

print(df.company_size.value_counts())   #Tres categorías

print(df.job_title.value_counts())   #Demasiadas, las dejaremos en seis + other
```

```python
coun=['US','GB','CA','DE','IN','FR','ES','GR']

lst=[]

for i in df.employee_residence:
    if i in coun:
        lst.append(i)
    else:
        lst.append('OTHER')
        
df.employee_residence=lst
```

```python
coun=['Data Scientist','Data Analyst','Machine Learning Engineer','Research Scientist']

lst=[]

for i in df.job_title:
    if i in coun:
        lst.append(i)
    else:
        lst.append('Other')
        
df.job_title=lst
```

## Modelos para 'salaries'

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X=Z.drop(columns='salary_in_usd')
y=Z.salary_in_usd

X_train, X_test, y_train, y_test = tts(X, y, train_size=0.8, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor as RFR  
from sklearn.metrics import mean_squared_error as mse

rfr=RFR()

rfr.fit(X_train, y_train)

mse(y_test, rfr.predict(X_test), squared=False)
```

### Random forest

```python
from sklearn.ensemble import RandomForestRegressor as RFR  
from sklearn.metrics import mean_squared_error as mse

rfr=RFR()

rfr.fit(X_train, y_train)

mse(y_test, rfr.predict(X_test), squared=False)
```

### Otros modelos

```python
from sklearn.tree import ExtraTreeRegressor as ETR
from sklearn.linear_model import LinearRegression as LinReg
from sklearn.linear_model import Lasso     
from sklearn.linear_model import Ridge        
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from lightgbm import LGBMRegressor as LGBMR


etr=ETR()
linreg=LinReg()
lasso=Lasso()
ridge=Ridge()
elastic=ElasticNet()
svr=SVR()
lgbmr=LGBMR()


modelos=[etr,linreg,lasso,ridge,elastic,svr,lgbmr]

for m in modelos:
    m.fit(X_train, y_train)
    print(m, mse(y_test,m.predict(X_test), squared=False))
```

```python
from xgboost import XGBRegressor as XGBR

xgbr=XGBR()

xgbr.fit(X_train, y_train)
mse(y_test,xgbr.predict(X_test), squared=False)
```

```python
from catboost import CatBoostRegressor as CTR

ctr=CTR()

ctr.fit(X_train, y_train)
mse(y_test,ctr.predict(X_test), squared=False)
```

```python
from sklearn.ensemble import GradientBoostingRegressor as GBR

gbr=GBR()

gbr.fit(X_train, y_train)
mse(y_test,gbr.predict(X_test), squared=False)
```
