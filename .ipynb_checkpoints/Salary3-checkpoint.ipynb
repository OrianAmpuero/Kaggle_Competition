```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data/salaries_data.csv')
```

```python
df
```

```python
#columna salary y salary_currency se eliminan al estar en la de salary in usd 

df.drop(columns=['salary','salary_currency'], inplace=True)

df
```

```python
Z=pd.get_dummies(df, columns=['experience_level','employment_type','job_title', 'employee_residence', 'company_location','company_size'], drop_first=False)
```

```python
Z.corr()
```

```python
Z.shape
```

```python
Z.info()
```

## Modelo

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X=Z.drop(columns='salary_in_usd')
y=Z.salary_in_usd

X_train, X_test, y_train, y_test = tts(X, y, train_size=0.8, test_size=0.2)
```

### Random forest

```python
from sklearn.ensemble import RandomForestRegressor as RFR  
from sklearn.metrics import mean_squared_error as mse

rfr=RFR()

rfr.fit(X_train, y_train)

y_pred=rfr.predict(X_test)

mse(y_test, y_pred, squared=False)
```

### Otros modelos

```python
from sklearn.tree import ExtraTreeRegressor as ETR
from sklearn.linear_model import LinearRegression as LinReg
from sklearn.linear_model import Lasso     
from sklearn.linear_model import Ridge        
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from lightgbm import LGBMRegressor as LGBMR


etr=ETR()
linreg=LinReg()
lasso=Lasso()
ridge=Ridge()
elastic=ElasticNet()
svr=SVR()
lgbmr=LGBMR()


modelos=[etr,linreg,lasso,ridge,elastic,svr,lgbmr]

for m in modelos:
    m.fit(X_train, y_train)
    print(m, mse(y_test,m.predict(X_test), squared=False))
```

```python
from xgboost import XGBRegressor as XGBR

xgbr=XGBR()

xgbr.fit(X_train, y_train)
mse(y_test,xgbr.predict(X_test), squared=False)
```

```python
from catboost import CatBoostRegressor as CTR

ctr=CTR()

ctr.fit(X_train, y_train)
mse(y_test,ctr.predict(X_test), squared=False)
```

```python
from sklearn.ensemble import GradientBoostingRegressor as GBR

gbr=GBR()

gbr.fit(X_train, y_train)
mse(y_test,gbr.predict(X_test), squared=False)
```
